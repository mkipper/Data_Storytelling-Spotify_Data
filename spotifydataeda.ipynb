```python
df_year = df.groupby('year')
```

```python
maximums = df_year.max()
maximums = maximums.reset_index()
print(maximums)
```

```python
minimums = df_year.min()
minimums = minimums.reset_index()
print(minimums)
```

```python
feature_list = 'bpm nrgy dnce val dur spch pop'.split()
max_min_yr = list()

for i in range(0, 10):
    for name in feature_list:
        max_min_yr.append(list(df_year)[i][1].sort_values(name).head(1))
        max_min_yr.append(list(df_year)[i][1].sort_values(name).tail(1))
```

```python
max_min_yr = pd.concat(max_min_yr)
max_min_yr
```

```python
max_min_yr.shape
```

```python
fig, ax = plt.subplots(figsize=(20,10))
sns.lineplot(data=maximums, x='year', y='bpm', color='black', marker='o')
sns.lineplot(data=minimums, x='year', y='bpm', color='tab:orange', marker='o')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
ax.set_xlabel('year', fontsize=30)
ax.set_ylabel('bpm', fontsize=30)
plt.legend(labels=['max', 'min'], fontsize=20)
plt.show()
```

Beats per minute (bpm) is the tempo of the song. The more beats per minute, the quicker, or more up-tempo, the song is. The overall trend over the 10 years of Spotify data for max song beats per minute is towards slower songs while min beats per minute trends towards quicker songs. Is this a 'regression to the mean'? 

```python
fig, ax = plt.subplots(figsize=(20,10))
sns.lineplot(data=maximums, x='year', y='nrgy', color='black', marker='o')
sns.lineplot(data=minimums, x='year', y='nrgy', color='tab:orange', marker='o')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
ax.set_xlabel('year', fontsize=30)
ax.set_ylabel('nrgy', fontsize=30)
plt.legend(labels=['max', 'min'], fontsize=20)
plt.show()
```

From this plot the max energy of hit songs on Spotify trends slightly downward from 2010 to 2019, although 2018-2019 starts to increase. While the min nrgy fluctuates wildly, 2017 and 2018 shows the min values levelling out and actually increasing slightly. It would be interesting to see where the trend is now with an extra 2 or 3 years of data.

```python
fig, ax = plt.subplots(figsize=(20,10))
sns.lineplot(data=maximums, x='year', y='dnce', color='black', marker='o')
sns.lineplot(data=minimums, x='year', y='dnce', color='tab:orange', marker='o')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
ax.set_xlabel('year', fontsize=30)
ax.set_ylabel('dnce', fontsize=30)
plt.legend(labels=['max', 'min'], fontsize=20)
plt.show()
```

Here, the max dance-iness of a song trends upwards over the 10 year span with occasional years going higher. Globally, people prefer dance-ier songs.

```python
fig, ax = plt.subplots(figsize=(20,10))
sns.lineplot(data=maximums, x='year', y='dur', color='black', marker='o')
sns.lineplot(data=minimums, x='year', y='dur', color='tab:orange', marker='o')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
ax.set_xlabel('year', fontsize=30)
ax.set_ylabel('dur', fontsize=30)
plt.legend(labels=['max', 'min'], fontsize=20)
plt.show()
```

The duration is how long the song is measured in seconds. The max duration in seconds seems to be all over the place! But the trend over 10 years, min and max values, leans towards shorter songs. 

```python
fig, ax = plt.subplots(figsize=(20,10))
sns.lineplot(data=maximums, x='year', y='spch', color='black', marker='o')
sns.lineplot(data=minimums, x='year', y='spch', color='tab:orange', marker='o')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
ax.set_xlabel('year', fontsize=30)
ax.set_ylabel('spch', fontsize=30)
plt.legend(labels=['max', 'min'], fontsize=20)
plt.show()
```

Spch is the Speechiness of a song - The higher the value the more spoken word the song contains. It seems the lowest value is 3. In 2012 there was a drastic dip in the max spch. I don't know how this is calculated, but this doesn't seem to be an error

```python
fig, ax = plt.subplots(figsize=(20,10))
sns.lineplot(data=maximums, x='year', y='pop', color='black', marker='o')
sns.lineplot(data=minimums, x='year', y='pop', color='tab:orange', marker='o')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
ax.set_xlabel('year', fontsize=30)
ax.set_ylabel('pop', fontsize=30)
plt.legend(labels=['max', 'min'], fontsize=20)
plt.show()
```

Pop measures the popularity of the song. This dataset implies a certain level of popularity so it's hard to imagine a song with '0' pop but yet is still in this data of 'Top Spotify songs from 2010-2019.' This occurs more than once so perhaps that value was missing in the 4 songs that registered '0' pop. I'm not sure how this is calculated. However, the trend over 10 years demonstrates that top songs are gaining in popularity

```python
fig, ax = plt.subplots(figsize=(20,10))
sns.lineplot(data=maximums, x='year', y='val', color='black', marker='o')
sns.lineplot(data=minimums, x='year', y='val', color='tab:orange', marker='o')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
ax.set_xlabel('year', fontsize=30)
ax.set_ylabel('val', fontsize=30)
plt.legend(labels=['max', 'min'], fontsize=20)
plt.show()
```

Val is the valence score of a song. This measures the positivity, or mood, of the song. The min values have not changed much during this timeframe. There was a spike in 2012. I'm curious if there were any political or social influences for this year. This data does collect information globally so it may difficult to determine the reason behind the spike. Globally, though, more people are preferring more positive songs.
